```python
import csv
import json
from datetime import datetime
import pandas as pd

# Define the CSV file and JSON file paths
csv_file = '41.18_Arrests_2013_-_2023.csv'
json_file = 'crime_typecasted.json'

# Define a function to type-cast CSV data
def typecast_csv_row(row):
    row['Report_ID'] = int(row['Report_ID'])
    row['Report_Type'] = str(row['Report_Type'])
    row['Year'] = int(row['Year'])
    
    # Check if the date string is not empty
    if row['Arrest_Date']:
        date_str = row['Arrest_Date'].strip()  # Remove leading/trailing spaces
        date_format = '%m/%d/%Y %H:%M' if len(date_str) > 10 else '%m/%d/%Y'
        
        # Convert the date string to a human-readable format
        row['Arrest_Date'] = datetime.strptime(date_str, date_format).strftime('%m/%d/%Y')
    
    row['Time'] = int(row['Time'])
    row['Time_Range'] = str(row['Time_Range'])
    row['Area_Name'] = str(row['Area_Name'])
    row['Reporting_District'] = int(row['Reporting_District'])
    row['Age'] = int(row['Age'])
    row['Sex'] = str(row['Sex'])
    row['Race'] = str(row['Race'])
    row['Charge_Group_Description'] = str(row['Charge_Group_Description'])
    row['Arrest_Type'] = str(row['Arrest_Type'])
    row['Charge'] = str(row['Charge'])
    row['Charge_Description'] = str(row['Charge_Description'])
    row['Disposition_Description'] = str(row['Disposition_Description'])
    row['Address'] = str(row['Address'])
    row['Cross_Street'] = str(row['Cross_Street'])
    row['LAT'] = float(row['LAT'])
    row['LON'] = float(row['LON'])
    row['CD'] = str(row['CD'])
    return row

data = []

with open(csv_file, 'r', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        # Check if any value in the row contains 'N/A'
        if 'N/A' not in row.values():
            row = typecast_csv_row(row)
            data.append(row)

# Create a DataFrame from the typecasted data
df = pd.DataFrame(data)

# Write the JSON data to a file
with open(json_file, 'w') as jsonfile:
    df.to_json(jsonfile, orient='records', lines=True)
```
